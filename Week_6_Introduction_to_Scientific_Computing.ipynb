## Explaining Pandas Data Structures

There are two types of data structures that are introduced with Pandas.

1. Series - A one-dimensional labeled array capable of holding any data type. The axis is refererd to as the index.

2. Dataframe - A 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a speradsheet or SQL table, or a dictionary of Series objects. It is generally the most common used pandas object.


Let's first explore a series.

## pandas Series

```python
#Let's grab a series of data from the dataframe above

TSLA_adjclose = TSLA_price_df['adjclose']
```

```python
type(TSLA_adjclose)
```

```python
TSLA_adjclose.head()
```

```python
#Let's do some cacluations with the series.
#We will need the help of numpy in this case.

import numpy as np

#This will calcuate returns for series
TSLA_adjclose_pct_change = TSLA_adjclose.pct_change()
TSLA_adjclose_pct_change.head()
```

```python
#This will calculate log returns for the series #log (p_1)-log(p_0) + log(p_2)-log(p_1)
TSLA_log_ret = np.log(TSLA_adjclose) - np.log(TSLA_adjclose.shift(1))
```

```python
TSLA_log_ret.head()
```

So why would you use a series instead of a dictionary?

When working with a pandas series, looping value-by-value is usually not necessary.

```python
# These are just for example purposes
TSLA_adjclose + TSLA_adjclose
```

```python
TSLA_adjclose*2
```

```python
np.exp(TSLA_adjclose)
```

# So everything we did with a series, you can apply to the dataframe.

Let's expand the used case a little bit, by making a dataframe with stock prices from 2 different companies

```python
tech_stocks = ['AAPL','MSFT']

techStocks_raw = YahooFinancials(tech_stocks)

techStocks_raw_data = techStocks_raw.get_historical_price_data('2023-01-01', '2023-12-31', 'daily')
```

```python
type(tech_stocks)
```

```python
print(techStocks_raw_data)
```

```python
#Now let's convert this data into a dataframe. Let's try a few ways.

techStocks_test1_df = pd.DataFrame.from_dict(techStocks_raw_data)

techStocks_test1_df.head()
```

```python
#So that doesn't give you the data the way you need.

techStocks_test2_df = pd.DataFrame.from_dict(techStocks_raw_data['AAPL']['prices'])
techStocks_test2_df.head()

#Let's just make 2 different dataframes

AAPL_price_df = pd.DataFrame.from_dict(techStocks_raw_data['AAPL']['prices'])
MSFT_price_df = pd.DataFrame.from_dict(techStocks_raw_data['MSFT']['prices'])
```

```python
#Convert the Indexes
AAPL_price_df.index = pd.to_datetime(AAPL_price_df['formatted_date'])
MSFT_price_df.index = pd.to_datetime(MSFT_price_df['formatted_date'])

AAPL_price_df.head()
```

```python
MSFT_price_df.head()
```

```python
#Now let's delete some columns in the dataframes
AAPL_price_df = AAPL_price_df.drop(columns=['date', 'formatted_date'])
```

```python
AAPL_price_df.head()
```

```python
MSFT_price_df = MSFT_price_df.drop(columns=['date', 'formatted_date'])
MSFT_price_df.head()
```

```python
#This is an example to show that you can do a whole function on the entire dataframe

#Let's create a dataframe so everything is a log return. You can remove the voulme column if you want.
AAPL_log_ret = np.log(AAPL_price_df) - np.log(AAPL_price_df.shift(1))
```

```python
AAPL_log_ret.head()
```

```python
#You can quickly get some descriptive statistics by telling pandas to describe the data to you.
AAPL_log_ret.describe()
```

```python
#Or use the same functions that a series has
AAPL_log_ret.mean()
```

```python
#if you just want one column you would simply
AAPL_log_ret['adjclose'].mean()
```

```python
#Same thing with plots
AAPL_price_df.plot()
```

## Question is now how do we create a correlation matrix between the 2 sets of data?

There are many ways of doing it. The way below is just one of them.

You could simply create a new dataframe from the dictionary by creating a new dictionary from the raw data.

This tests your knowledge of data structures.

```python
AAPL_adjclose = AAPL_price_df['adjclose']
MSFT_adjclose = MSFT_price_df['adjclose']
```

```python
MSFT_adjclose.head()
```

```python
techStock_adjclose = pd.DataFrame({'AAPL':AAPL_adjclose, 'MSFT': MSFT_adjclose})
```

```python
techStock_adjclose.head()
```

```python
techStock_log_ret = np.log(techStock_adjclose) - np.log(techStock_adjclose.shift(1))
```

```python
techStock_log_ret.head()
```

```python
#Simply get the correlation matrix
techStock_log_ret.corr()
```
