```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = -x  # Perfect negative correlation

# Plot the scatter graph
plt.scatter(x, y, color='blue', label='y = -x')
plt.title('Perfectly Negative Correlated Scatter Graph')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.scatter(x, y, color='red', label='Asset A vs Asset B')
plt.title('Perfectly Negative Correlation Between Two Assets')
plt.xlabel('Return of Asset A')
plt.ylabel('Return of Asset B')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
