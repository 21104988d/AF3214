```python
import requests
from bs4 import BeautifulSoup

Sp500Wiki = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
spResponse = requests.get(Sp500Wiki)
soup = BeautifulSoup(spResponse.text)

sp500_rows = soup.table.find_all('tr')
sp500Stocks = []

for row in sp500_rows:
    if row.td == None:
        continue
    else:
        print (row.find_all('td')[1].a.text)
        sp500Stocks.append(row.td.a.text)

```

```python
import random

random_stock = []
for i in range(3):
    random_stock.append(random.choice(sp500Stocks))
print(random_stock)
```

```python
from alpha_vantage.timeseries import TimeSeries

ts = TimeSeries(key='HNQ62X7XN1BICIIA', output_format='pandas')

data, meta_data = ts.get_daily(symbol=random_stock[0], outputsize='full')
data1, meta_data1 = ts.get_daily(symbol=random_stock[1], outputsize='full')
data2, meta_data2 = ts.get_daily(symbol=random_stock[2], outputsize='full')

sort_data = data.sort_index()
sort_data1 = data1.sort_index()
sort_data2 = data2.sort_index()

stock1 = sort_data.loc['2025-04-01':'2025-04-30']
stock2 = sort_data1.loc['2025-04-01':'2025-04-30']
stock3 = sort_data2.loc['2025-04-01':'2025-04-30']
```

```python
import numpy as np
import pandas as pd

weight = np.array([1/3, 1/3, 1/3])
portfolio = pd.DataFrame({'stock1': stock1['4. close'], 
                          'stock2': stock2['4. close'], 
                          'stock3': stock3['4. close']})

annulized_cov_matrix = portfolio.cov() * 252
portfolio_variance = np.dot(weight.T, np.dot(annulized_cov_matrix, weight))

portfolio_variance2 = weight @ annulized_cov_matrix @ weight
print(portfolio_variance2)
```

```python
monthly_return = np.sum(portfolio.mean() * weight) * 22
monthly_cov_matrix = portfolio.cov() * 22
monthly_variance = np.dot(weight.T, np.dot(monthly_cov_matrix, weight))
monthly_std = np.sqrt(monthly_variance)
```

```python
pdf_list = ['ssrn-1150076.pdf', 'ssrn-1171102.pdf', 'ssrn-3694637.pdf', 'ssrn-4964419.pdf']
random_pdf = random.choice(pdf_list)
```

```python
import PyPDF2

pdf_reader = PyPDF2.PdfReader(random_pdf)
number_of_pages = len(pdf_reader.pages)
random_page = random.randint(0, number_of_pages - 1)
page = pdf_reader.pages[random_page]
page_text = page.extract_text()
print(page_text)
```

```python
import textstat

readability_score = textstat.gunning_fog(page_text)
print(readability_score)
```
